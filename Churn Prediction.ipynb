```python
df["TotalCharges"]=pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df["Churn"]=df.Churn.replace("No",0)
df["Churn"]=df.Churn.replace("Yes",1)
```

```python
df.tenure.astype("float64", errors="ignore")
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.drop("customerID",axis=1,inplace=True)
```

```python
cols=['gender', 'SeniorCitizen', 'Partner', 'Dependents',
     'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod']
```

```python
df1=pd.get_dummies(df,columns=cols ,drop_first=True)
df.drop(cols, axis=1,inplace=True)
df=pd.concat([df1,df],axis=1)
df=df.loc[:,~df.columns.duplicated()]
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
cols2=["MonthlyCharges","TotalCharges","tenure"]
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
df[cols2]=scaler.fit_transform(df[cols2])
```

```python
df.head()
```

```python
X= df.drop('Churn',axis=1)
y=df.Churn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model= keras.Sequential([
    keras.layers.Dense(units=15, kernel_initializer="he_uniform",activation='relu', input_dim=30),
    keras.layers.Dense(units=6, kernel_initializer="he_uniform",activation='relu'),
    keras.layers.Dense(units=1, kernel_initializer="he_uniform",activation='sigmoid')
        

])
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train, y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_p= model.predict(X_test)
```

```python
y_p
```

```python
y_pred=[]
for i in y_p:
    if i>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
